```python
import os
import sys
import random
import numpy as np
import matplotlib.pyplot as plt
import skimage
from skimage.io import imread, imshow, imsave
import cv2
import pandas as pd
from keras_retinanet.utils.image import TransformParameters, adjust_transform_for_image, apply_transform,read_image_bgr
from PIL import Image
```

# Image Augmentation

```python
path = r'C:\Users\Asus\Desktop\PG Project\Nandhini\Phase 2\Dataset\Kvasir-SEG\images\cju1h89h6xbnx08352k2790o9.jpg'
x =np.ascontiguousarray(Image.open(path).convert('RGB'))
```

```python
plt.axis('off')
imshow(x)
plt.savefig(r"C:\Users\Asus\Desktop\image.png", dpi=600)
```

## Rotation

```python
param = TransformParameters()
```

```python
rotation_matrix = rotation(0.5) #radian
transform = adjust_transform_for_image(rotation_matrix, x, True)
image_rotation = apply_transform(transform, x, param)
```

```python
plt.axis('off')
imshow(image_rotation)
plt.savefig(r"C:\Users\Asus\Desktop\image-rotation.png", dpi=600)
```

## Scaling

```python
scaling_matrix = random_scaling((0.8,0.8),(1.3,1.3)) 
transform_scaling = adjust_transform_for_image(scaling_matrix, x, True)
image_scaling = apply_transform(transform_scaling, x, param)
```

```python
plt.axis('off')
imshow(image_scaling)
plt.savefig(r"C:\Users\Asus\Desktop\image-scaling.png", dpi=600)
```

## Flip x

```python
flip_matrix = random_flip(0,1)
transform_flip = adjust_transform_for_image(flip_matrix, x, True)
image_flip = apply_transform(transform_flip, x, param)
```

```python
plt.axis('off')
imshow(image_flip)
plt.savefig(r"C:\Users\Asus\Desktop\image-flip-x.png", dpi=600)
```

## Flip Y

```python
flip_matrix_y = random_flip(1,1)
transform_flip_y = adjust_transform_for_image(flip_matrix_y, x, True)
image_flip_y = apply_transform(transform_flip_y, x, param)
```

```python
plt.axis('off')
imshow(image_flip_y)
plt.savefig(r"C:\Users\Asus\Desktop\image-flip-y.png", dpi=600)
```

## Contrast

```python
image_contrast = adjust_contrast(x, 1.2)
```

```python
plt.axis('off')
imshow(image_contrast)
plt.savefig(r"C:\Users\Asus\Desktop\image-contrast.png", dpi=600)
```

## Brightness

```python
image_brightness = adjust_brightness(x, -0.1)
```

```python
plt.axis('off')
imshow(image_brightness)
plt.savefig(r"C:\Users\Asus\Desktop\image-brightness.png", dpi=600)
```
